```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Preprocessing Input data
data = pd.read_csv('Housing.csv')
data.head()
```

```python
corr_matrix=data.corr()
print(corr_matrix)
```

```python
X = data.iloc[:, 0].values
Y = data.iloc[:, -1].values
plt.scatter(X, Y)
plt.ylabel('price')
plt.xlabel('plotsize')
plt.show()
#print(Y)
```

```python
# Building the model
m = 0
c = 0

L = 0.0001  # The learning Rate
iterations = 49  # The number of iterations to perform gradient descent

n = float(len(X)) 


# Gradient Descent 
for i in range(iterations): 
    Y_pred = m*X + c  
    Deriv_m = (-2/n) * sum(X * (Y - Y_pred))  
    Deriv_c = (-2/n) * sum(Y - Y_pred)  
    m = m - L * Deriv_m  
    c = c - L * Deriv_c
    
print (m, c)
```

```python
# Making predictions
Y_pred = m*X + c

plt.scatter(X, Y)
plt.scatter(X, Y_pred)
plt.plot([min(X), max(X)], [min(Y_pred), max(Y_pred)], color='red') # predicted
plt.show()
```
